## Train Test Split

```python
from sklearn.model_selection import train_test_split

x= cardio.drop(columns='cardio')
y= np.array(cardio['cardio'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
#Scaling

from sklearn.preprocessing import StandardScaler

scaler= StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
print(X_train.shape)
print(X_test.shape)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression


lr= LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_pred_lr=lr.predict(X_test)
y_pred_prob_lr=lr.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score,roc_auc_score,classification_report,roc_curve


print("Accuracy_score:", accuracy_score(y_test,y_pred_lr))
print("Roc_auc_score:", roc_auc_score(y_test,y_pred_lr))
print("Classification_report:\n", classification_report(y_test,y_pred_lr))
```

```python
fpr,tpr,threshold=roc_curve(y_test, y_pred_prob_lr[:,1])

plt.plot(fpr,tpr)
plt.title('ROC_AUC_Curve')
plt.xlabel("False Positive Rate")
plt.ylabel('True Positive Rate')
plt.grid()
plt.show()
```

```python
from xgboost import XGBClassifier

from sklearn.model_selection import GridSearchCV
```

```python

param_grid = {
    'n_estimators': [10,50,100, 200],
    'max_depth': [3, 5, 7,10],
    'learning_rate': [0.01,0.05, 0.1, 0.2],
    'subsample': [0.8, 1.0]
}

grid_search = GridSearchCV(XGBClassifier(), param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train,y_train)
```

```python
print("Best Parameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)
```

```python
y_pred_xg=grid_search.predict(X_test)
y_pred_prob_xg=grid_search.predict_proba(X_test)
```

```python
print("Accuracy_score:", accuracy_score(y_test,y_pred_xg))
print("Roc_auc_score:", roc_auc_score(y_test,y_pred_xg))
print("Classification_report:\n", classification_report(y_test,y_pred_xg))
```

```python
fpr,tpr,threshold=roc_curve(y_test, y_pred_prob_xg[:,1])

plt.plot(fpr,tpr)
plt.title('ROC_AUC_Curve')
plt.xlabel("False Positive Rate")
plt.ylabel('True Positive Rate')
plt.grid()
plt.show()
```
